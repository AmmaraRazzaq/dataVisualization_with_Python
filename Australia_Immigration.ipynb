### Scatter Plots

```python
df_tot.index = map(int, df_tot.index)
df_tot.head()
```

```python
df_tot.drop(['level_0'], axis=1, inplace=True)
df_tot.head()
```

```python
df_tot.columns = ['year','total']
```

```python
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10,6), color='darkblue')

plt.title('Total Immigration to Australia from 1980 - 2008')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
```

```python
#Analyze the trend using a regression line
x = df_tot['year']
y = df_tot['total']

fit = np.polyfit(x,y,deg=1)
fit
```

```python
#plot the regression line on the scatter plot
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10,6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2008')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

#plot regression line
plt.plot(x, fit[0]*x+fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))
```

### Bubble Plot

```python
df.head(3)
```

```python
df_t = df[years].transpose()
df_t.head(3)
```

```python
df_t.index = map(int, df_t.index)
```

```python
df_t.index.name = 'Year'
df_t.head(3)
```

```python
df_t.reset_index(inplace=True)
df_t.head(3)
```

```python
#Create the normalized weights
df_t.rename(columns={'United Kingdom of Great Britain and Northern Ireland':'UK','United States of America':'USA'}, inplace=True)
df_t.head(3)
```

```python
# (x - xmin)/(xmax - xmin)
norm_UK = (df_t['UK'] - df_t['UK'].min()) / (df_t['UK'].max() - df_t['UK'].min())

norm_USA = (df_t['USA'] - df_t['USA'].min()) / (df_t['USA'].max() - df_t['USA'].min())
```

```python
ax0 = df_t.plot(kind='scatter', x='Year', y='UK', figsize=(14,8), alpha=0.5, color='green', s=norm_UK*1000+10, xlim=(1980,2008))
ax1 = df_t.plot(kind='scatter', x='Year', y='USA', alpha=0.5, color='blue', s=norm_USA*1000+10, ax=ax0)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from UK and USA from 1980 - 2008')
ax0.legend(['UK', 'USA'], loc='upper left', fontsize='x-large')
```
