```python
#drop all rows of emigrants    #df['condition'] = new dataframe
df = df[df['Type'] == 'Immigrants']
df.head(2)
```

```python
df.tail()
```

```python
#drop unnecessary columns
df.drop(['AREA','REG','DEV','Type','Coverage',2009,2010,2011,2012,2013], axis=1, inplace=True)
df.head(2)
```

```python
#rename the columns so that they make sense
df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
```

```python
#add a Total column that sums up the total immigrants by country over the entire period
df['Total'] = df.sum(axis=1)
df.head(2)
```
