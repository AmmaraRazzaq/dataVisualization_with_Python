```python
#drop all rows of emigrants    #df['condition'] = new dataframe
df = df[df['Type'] == 'Immigrants']
df.head(2)
```

```python
df.tail()
```

```python
#drop unnecessary columns
df.drop(['AREA','REG','DEV','Type','Coverage',2009,2010,2011,2012,2013], axis=1, inplace=True)
df.head(2)
```

```python
#rename the columns so that they make sense
df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
```

```python
#add a Total column that sums up the total immigrants by country over the entire period
df['Total'] = df.sum(axis=1)
df.head(2)
```

```python
#check how many null objects we have in data set
df.isnull().sum()
```

```python
df['Country']
```

```python
#set country column as index to make querying easier
df.set_index('Country', inplace=True)
df.head(2)
```

```python
#remove the name of index -> Optional
df.index.name = None
```

### Produce Line Plot

```python
#Compare the trend of top 5 countries that contributed the most to immigration to Australia
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df.head()
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.plot(kind='line', figsize=(14,8))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

### Produce Area Plot

```python
#Area plots are stacked by default
df_top5.plot(kind='area',stacked=False, figsize=(20,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

### Plot Histograms

```python
#frequency distribution of the number of immigrants in 2008 from various countries to Australia.
df['2008'].head()
```

```python
count, bins = np.histogram(df['2008'])
print(count)
print(bins)
```

```python
df['2008'].plot(kind='hist', figsize=(8,5), xticks = bins)

plt.title('Histogram of Immigration from 195 Countries in 2008') 
plt.ylabel('Number of Countries') 
plt.xlabel('Number of Immigrants') 
```

```python
# Immigration distribution for Pakistan, India, Bangladesh over the years
df.loc[['Pakistan','India','Bangladesh'],years]
```

```python
df_t = df.loc[['Pakistan','India','Bangladesh'],years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist',figsize=(10,6))

plt.title('Histogram of Immigration from Pakistan, India and Bangladesh from 1980 - 2008')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

```python
count, bins_edges = np.histogram(df_t,15)
```

```python
df_t.plot(kind='hist', figsize=(10,6), bins=15, alpha=0.6, xticks=bins_edges, color=['coral', 'darkslateblue', 'mediumseagreen'])
plt.title('Histogram of Immigration from Pakistan, India and Bangladesh from 1980 - 2008')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

```python
xmin = bins_edges[0] - 5
xmax = bins_edges[-1] + 5

df_t.plot(kind='hist',
          figsize=(10,6),
          bins=15,
          xticks=bins_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin,xmax))
plt.title('Histogram of Immigration from Pakistan, India and Bangladesh from 1980 - 2008')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```
